```python
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys, os
from scipy.io import wavfile
import librosa
import librosa.display

FNAME = "subvocal_12_24_2"

ROOT = os.getcwd()
AUD_ROOT = os.path.join(*[ROOT, "openbci_data", "openbci_to_wav", FNAME, 'split_audio'])
IMG_ROOT = os.path.join(*[ROOT, "images_scaled", FNAME])

print('-' * 42)
print("Current Directory:       {}".format(ROOT))
print("Current AUDIO Directory: {}".format(AUD_ROOT))
print("CUrrent IMG Directory:   {}".format(IMG_ROOT))
print('-' * 42)
```

```python
def plot_wav(channel, fname):
    spf = wave.open(os.path.join(*[AUD_ROOT, 
                                   'lights-on',
                                   channel,fname]),'r')
    spf2 = wave.open(os.path.join(*[AUD_ROOT, 
                                   'turn-off',
                                   channel,fname]),'r')
    plt.figure(figsize=(6,2))
    #Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.fromstring(signal, 'Int16')
    fs = spf.getframerate()
    
    signal2 = spf2.readframes(-1)
    signal2 = np.fromstring(signal2, 'Int16')
    fs2 = spf2.getframerate()
    #If Stereo
    
    if spf.getnchannels() == 2:
        print('Just mono files')
        sys.exit(0)
    time=np.linspace(0, len(signal)/fs, num=len(signal))
    time2=np.linspace(0, len(signal)/fs2, num=len(signal2))
    
    ax1 = plt.subplot2grid((1,2), (0,0), rowspan=1, colspan=1)
    ax2 = plt.subplot2grid((1,2), (0,1), rowspan=1, colspan=1)
    plt.axis('off')
    plt.title(fname)
    ax1.plot(time,signal)
    plt.axis('off')
    ax2.plot(time2,signal2)
    plt.show()
```

```python
for i in range(10):
    plot_wav('ch2', '0000' + str(i) + '.wav')
```
