```python
import pandas as pd
import numpy as np

KEY = "test_key.csv"
USER = "rail_test.csv"
MAPPING = False
mapping = {"one": 1, "two": 2}

key = pd.read_csv(KEY)
user = pd.read_csv(USER)
key = key.drop(key.columns[0], axis=1)

if MAPPING == True:
    user['label'] = user['label'].map(mapping)
result = key.sort_values('fileName')
result = result.reindex()
result = result.reset_index()
user = user.sort_values('filename')
user = user.reset_index()
result['user'] = user['label']
result['predict'] = result.apply(lambda x: 'O' if x['label'] == x['user'] else 'X', axis=1)
predict = result['predict'].value_counts()
print("="*42)
print("ACCURACY : {:0.2f} %".format(predict['O'] / 1002 * 100))
print("="*42)
```
